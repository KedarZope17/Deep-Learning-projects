```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
TataGlobal_train = pd.read_csv('NSE-TATAGLOBAL.csv')
TataGlobal_train
```

```python
TataGlobal_train.iloc[:, 1:2]
```

```python
Train = TataGlobal_train.iloc[:, 1:2].values
Train
```

```python
from sklearn.preprocessing import MinMaxScaler
Scalar = MinMaxScaler(feature_range = (0, 1))
Train_scaled = Scalar.fit_transform(Train)
Train_scaled
```

```python
X_train = [] #independent
Y_train = [] #dependent
for i in range(60, 2035): #Creating a Data Structure with 60 time steps and 1 output 
    X_train.append(Train_scaled[i-60:i, 0])#index range(0:59,1:60,2:61,...)
    Y_train.append(Train_scaled[i, 0])#index (60,61,62,...)
    print(f"x_train index {i-60} to {i}")
    print(f"y_train index {i}")
    print(f"x_train values {Train_scaled[i-60:i, 0]}")
    print(f"y_train values {Train_scaled[i, 0]}")
    #print(X_train,y_train)
X_train, Y_train = np.array(X_train), np.array(Y_train)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1)) #reshaping X_train from 2D to 3D
X_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout #regularization technique for better learning
```

```python
regressor = Sequential() #Initialization
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1))) #layer addition
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True)) #2nd layer
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True)) #3rd layer
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50)) #4th layer
regressor.add(Dropout(0.2))
regressor.add(Dense(units = 1)) #output layer
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error') #Compilation
# Fitting the RNN to the Training set
```

```python
regressor.fit(X_train, Y_train, epochs = 200, batch_size = 16) #training
```

```python
X_test = pd.read_csv('test_data.csv') #Test dataset
X_test
```

```python
X_test = X_test.drop(['Unnamed: 0'],axis=1)
X_test = X_test.values
X_test
```

```python
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1],1)) #reshaping from 2d to 3d
```

```python
predicted_stock_price = regressor.predict(X_test) #o/p is in rangeof 0-1
predicted_stock_price = Scalar.inverse_transform(predicted_stock_price)
```

```python
real_stock_price = pd.read_csv('real_stock_price.csv') #actual price dataset
real_stock_price
```

```python
real_stock_price = real_stock_price['0'].values
real_stock_price
```

```python
plt.plot(real_stock_price, color = 'red', label = 'Real TATA Stock Price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted TATA Stock Price')
plt.title('TATA Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('TATA Stock Price')
plt.legend()
plt.show()
```
